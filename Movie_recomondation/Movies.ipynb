```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python
def preproces_text(text):
  text = str(text).lower()
  text = re.sub(r"[^a-zA-Z\s]",'',text)

  words = text.split()
  words = [word for word in words if word not in stop_words]

  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)



```

```python
df["tags"] = df["tags"].apply(preproces_text)
```

```python
df = df.reset_index(drop =True)
```

```python
indices = pd.Series(df.index, index=df["title"]).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,stop_words="english",ngram_range=(1,2))
```

```python
tfisf_matrix = tfidf.fit_transform(df["tags"])
```

```python
tfisf_matrix.shape
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
  if title not in indices:
    return ["Movie not found"]

  idx = indices[title]
  sim_score = cosine_similarity(tfisf_matrix[idx],tfisf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+  1]
  return df["title"].iloc[similar_idx]
```

```python
recommend("Avatar")
```

```python
import pickle
pickle.dump(tfisf_matrix,open('tfidf_matrix.pkl','wb'))

pickle.dump(indices,open('indices.pkl','wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))

```
